# Using hls4ml to generate hdl that we will deploy

```python
import hls4ml
import qkeras
from qkeras.utils import load_qmodel
from tensorflow.keras.models import load_model
```

```python
# Load saved keras model
model = load_qmodel("./src/q_auto_encoder.keras")
```

```python
config = hls4ml.utils.config_from_keras_model(
     model,
     granularity='name',
     default_precision='fixed<8,1>',
     backend='Vitis')
```

```python
hls_model = hls4ml.converters.convert_from_keras_model(
      model,
      hls_config=config,
      output_dir="./src",
      io_type='io_stream',
      backend='Vitis'
  )
```
